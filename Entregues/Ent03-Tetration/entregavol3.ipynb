```python
#INPUT: n: number of iterates
#       z: value to compute de sequence z^(z^(z^...))
#OUPUT: z^(z^(z^...)) (with n-1 terms)
def f(n,z):
    zn=z.n()
    for i in range(0,n):
        zn=(e^(zn*log(z))).n()
    return zn
```

```python
def iterats(n,z):
    return [f(i,z) for i in range(0,n+1)]
def iterats_limit(z):
    L=[z.n(),(z^z).n()]
    i=0
    while abs(L[i]-L[i+1])>1.e-20:
        L.append(f(i+2,z))
        i=i+1
    return L
```
